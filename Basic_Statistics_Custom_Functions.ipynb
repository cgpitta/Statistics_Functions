#### 4 types of data in statistics:

	• Nominal Data
		○ Labels; no quantitative value; can be grouped (gender, phone numbers)
	• Ordinal data
		○ Non-numerical data; can be ranked (pain measurement scale) but the distance between the numbers might not be the same
	• Interval Data
		○ Numerical values; equal distance between; known order and differences
	• Ratio Data
		○ Can be compared

#### Statistics

	• Descriptive - Summarizing Data
	• Inferential - making inferences and determining relationships

Population in statistics means the entire set (of interest)

A sample is a subset of a population and has to be randomly selected
Random means all items have equal chance to be selected

Central tendency means a value describes the center or central location of a data set

Distribution: shows all values in a data set and their frequency



#### When to use Mode, Mean, or median?

	◊ Mode: nominal data; outliers are fine. Determining which brands for example?

	◊ Mean: Interval and ratio data not excessively skewed. What is the avg. salary? 
	
	◊ Median: Ordinal Data - skewed data is fine. How satisfied are you?




```python
#population standard deviation
def stdev_pop(data):
    N=len(data)
    mean = sum(data)/N
    sum_of_squares=0

    for val in data:
        sum_of_squares += (val-mean)**2

    var = sum_of_squares/N #population variance

    return var ** 0.5


def stdev_s(data):
    N=len(data)
    mean = sum(data)/N
    sum_of_squares=0

    for val in data:
        sum_of_squares += (val-mean)**2

    var = sum_of_squares/(N-1)  ##sample variance

    return var ** 0.5
```

```python
#z-score population or sample

def z_score(value,mean,std_dev):
    return (value-mean)/std_dev
    
```

```python
def empiracle_rule(data):
	try:
		datum = list(data)

		N = len(datum)

		if N == 0:
			return None

		#calculate mean

		mean = sum(datum)/N


		#calc standard deviation
		sum_of_squares = 0

		for val in datum:
			sum_of_squares += (val-mean)**2

		std_dev = (sum_of_squares/N)**2 # population standard deviation



	# count points within 1, 2, and 3 standard deviations
		count_std_1=0
		count_std_2=0
		count_std_3=0

		for x in datum:
		
			if abs(x-mean)<= std_dev:
			count_std_1+=1

			if abs(x-mean) <= (2*std_dev):
			count_std_2 +=1
		
			if abs(x-mean) <= (3*std_dev):
			count_std_3 +=1


		within_1std = (count_std_1/N) *100
		within_2std = (count_std_2/N)*100
		within_3std = (count_std_3/N)*100


		return within_1std,within_2std,within_3std
```

```python
def mode(x):
    frequency = []

    for val in x:
        if val in frequency:
            frequency[val] += 1
        else:
            frequency[val] = 1

    max_count = max(frequency.values())

    #collecgt all the values with the max frequency
    mode = [key for key, count in frequency.items() if count == max_count]


    return mode[0] if len(mode) == 1 else mode
```

```python
def covariance(x,y):
    meanx = sum(x)/len(x)
    meany = sum(y)/len(y)
    n = len(x) + len(y)

    covariance = sum((xi - meanx) * sum(yi-meany) for xi,yi in zip(x,y))

    return covariance/n

def covariances(x,y):
    meanx = sum(x)/len(x)
    meany = sum(y)/len(y)
    n = len(x) + len(y)

    covariance = sum((xi - meanx) * sum(yi-meany) for xi,yi in zip(x,y))

    return covariance/(n-1)
```
